```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("gsearch_jobs.csv")
```

```python
# Preview first 5 rows
df.head()
```

```python
# Shape of data
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

```python
# Check column names and data types
df.info()
```

```python
df.isnull().sum()
#check missing values in each column
```

```python
#STEP -2 DATA CLEANING 
# Drop unnecessary columns
df_clean = df.drop(columns=[
    'Unnamed: 0', 'index', 'thumbnail', 'commute_time', 
    'extensions', 'job_id', 'description_tokens'
])
```

```python
df_clean.shape[1]
```

```python
# Fill small missing text columns with 'Unknown'
df_clean['location'] = df_clean['location'].fillna('Unknown')
df_clean['via'] = df_clean['via'].fillna('Unknown')
df_clean['schedule_type'] = df_clean['schedule_type'].fillna('Unknown')
df_clean['work_from_home'] = df_clean['work_from_home'].fillna('Unknown')
```

```python
# Convert date columns to datetime
# df_clean['posted_at'] = pd.to_datetime(df_clean['posted_at'], errors='coerce')
df_clean['date_time'] = pd.to_datetime(df_clean['date_time'], errors='coerce')
```

```python
# Strip extra spaces in key columns
df_clean['title'] = df_clean['title'].str.strip()
df_clean['company_name'] = df_clean['company_name'].str.strip()
```

```python
# Check results after data cleaning and preparation 
df_clean.info()
```

```python
# Remove 'posted_at' column because it contains non-standard text values 
# like '12 hours ago', which cannot be converted to proper date/time format.
# It‚Äôs not useful for analysis in this form.
df_clean.drop(columns=['posted_at'], inplace=True)
```

```python
#KPI's CREATION 
```

```python
# KPI 1 ‚Äî Total Job Listings
total_jobs = df_clean.shape[0]
print("üìä Total Job Listings:", total_jobs)
```

```python
# KPI 2 ‚Äî Top 10 Companies Hiring
top_companies = df_clean['company_name'].value_counts().head(10).reset_index()
top_companies.columns = ['Company Name', 'Job Listings']

print("üè¢ Top 10 Companies Hiring:")
display(top_companies)
```

```python
# KPI 3 ‚Äî Top 10 Job Titles
top_titles = df_clean['title'].value_counts().head(10).reset_index()
top_titles.columns = ['Job Title', 'Job Listings']

print("üíº Top 10 Job Titles:")
display(top_titles)
```

```python
# KPI 4 ‚Äî Remote Jobs %
total_jobs = len(df_clean)
df_clean['work_from_home'] = df_clean['work_from_home'].astype(str)

remote_jobs = df_clean[df_clean['work_from_home'].str.lower() == 'true']
remote_job_count = len(remote_jobs)
remote_job_percent = (remote_job_count / total_jobs) * 100

print(f"üè† Remote Jobs: {remote_job_count} out of {total_jobs} ({remote_job_percent:.2f}%)")
```

```python
# KPI 5 ‚Äî Average Salary (if available)
avg_salary = df_clean['salary_avg'].mean()
print(f"üí∞ Average Salary (across listings with data): ${avg_salary:,.2f}")
```

```python
# KPI 6 ‚Äî Jobs by Schedule Type
jobs_by_schedule = df_clean['schedule_type'].value_counts().reset_index()
jobs_by_schedule.columns = ['Schedule Type', 'Job Listings']

print("üïí Jobs by Schedule Type:")
display(jobs_by_schedule)
```

```python
# KPI 7 ‚Äî Job Posting Trend Over Time
df_clean['month'] = df_clean['date_time'].dt.to_period('M')
trend_over_time = df_clean.groupby('month').size().reset_index(name='job_listings')

print("üìÖ Job Posting Trend (Monthly):")
display(trend_over_time)
```

```python
# ‚úÖ Create KPI Summary Table

kpi_summary = pd.DataFrame({
    "KPI": [
        "Total Job Listings",
        "Top 10 Companies Hiring",
        "Top 10 Job Titles",
        "Remote Jobs %",
        "Average Salary (from salary_avg)",
        "Jobs by Schedule Type",
        "Job Posting Trend Over Time"
    ],
    "Result": [
        f"{df_clean.shape[0]:,}",
        ", ".join(top_companies['Company Name'].tolist()),
        ", ".join(top_titles['Job Title'].tolist()),
        f"{remote_job_percent:.2f}%",
        f"${avg_salary:,.2f}",
        "Full-time, Contractor, Temp Work lead the market",
        "Job postings peaked around Jan 2023 and dipped mid-2024, showing recovery in early 2025"
    ]
})

display(kpi_summary)
```

```python
# Save KPI summary to CSV for Power BI import
kpi_summary.to_csv("kpi_summary.csv", index=False)
```
